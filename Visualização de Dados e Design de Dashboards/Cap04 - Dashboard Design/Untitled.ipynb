```python
# Imports
import json
import requests
import http.client
import pandas as pd
import plotly
import plotly.express as px
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore", category = FutureWarning)
```

## Análise de Dados Covid-19 

```python
# Conexão aos dados em tempo real via API
# https://coronavirus.jhu.edu/map.html
# https://covid19api.com/
conn = http.client.HTTPSConnection("api.covid19api.com")
```

```python
# Fazendo requisição dos dados sumarizadis e gravando em um objeto json
payload = ''
headers = {}
conn.request("GET","/summary",payload,headers)
res = conn.getresponse()
data  = res.read().decode('UTF-8')
covid = json.loads(data)
```

```python
# Converte os dados para formato dataframe
df = pd.DataFrame(covid['Countries'])
```

```python
df
```

```python
# Obtendo dados da população total de cada pais - Leitura arquivo excel
df1 = pd.read_excel('Dados\Populacao.xlsx', usecols = 'C,D' )
```

```python
df1
```

```python
#Fazendo merge dos dataframe, considerando apenas países com correspondência nas duas tabelas
df = df.merge(df1, how='inner', on='Country')
```

```python
df
```

```python
# Remove as colunas não utilizadas
covid = df.drop(columns = ['ID', 'Slug', 'NewConfirmed', 'NewDeaths', 'NewRecovered', 'Premium', 'Date', 'TotalRecovered'], axis = 1)
```

```python
covid
```

```python
# Calculando o total de Sobreviventes
# Obs: O total 
covid['Survivors'] = covid['TotalConfirmed'] - covid['TotalDeaths']
```

```python
# Calculando a porcentagem de sobreviventes 
covid['(%) Survivors'] = ((covid['Survivors'] / covid['TotalConfirmed']) * 100).round(2)
```

```python
# Calculando a porcentagem de mortes
covid['(%) Deaths'] = 100 - covid['(%) Survivors']
```

```python
# Calculando a taxa de mortes por 1 milhão de habitantes
covid['deathRate'] = ((covid['TotalDeaths']/covid['Population'])*1000000).round(0)
```
